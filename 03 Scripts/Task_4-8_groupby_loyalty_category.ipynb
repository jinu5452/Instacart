### Step 4. creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']>40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order']>10), 'loyalty_flag' ] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order']<= 10, 'loyalty_flag']='New customer'
```

### Step 5. difference between the spending habits of the three types of customers

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

Loyal customer's average spend is lower than Regular and New customer types. The hiesst spend by customer type is New customer. 

### Step 6. different types of spenders. 10>= : high spender / 10< : Low spender. 

```python
# each user's average price across all their orders, spending flag

ords_prods_merge['average_spending']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.tail()
```

```python
ords_prods_merge['sum_spending']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.sum)
```

```python
ords_prods_merge.tail()
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spending']>=10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.tail()
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

### Step 7. frequent versus non-frequent customers, creating order frequency flag, according to the median in the “days_since_prior_order”

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
ords_prods_merge.tail()
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency']>20, 'frequency_flag'] = 'Non-frequency customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency']<=20) & (ords_prods_merge['order_frequency'] > 10) , 'frequency_flag'] = 'RegularF customer'
```

```python
ords_prods_merge.tail()
```
