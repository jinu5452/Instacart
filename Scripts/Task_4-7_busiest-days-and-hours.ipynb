```python
# define funtion price_label(row)

def price_label(row):
    if row['prices'] <= 5: 
        return 'Low-range product'
    elif (row['prices']  > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

### Step 2. 2 busiest days and 2 slowest days

```python
# 2 busiest days - 0 and 1 // slowest days - 3 and 4

busyslow =[]
for days in ords_prods_merge["order_day_of_week"]:
  if days == 0 or days==1 :
    busyslow.append("Busiest days")
  elif days == 4 or days==3:
    busyslow.append("Slowest days")
  else:
    busyslow.append("Regularly busy")
```

### Step 4 identify the busiest hours of the day 
 - labeled Most orders, Average orders, and Fewest orders. 
 - Create a new column containing these labels called busiest_period_of_day.

```python
# check frequency of order by hours
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

#### criteria for categorizeing is not given. only 'too many users'
 - most orders: more than 2,500,000 (10, 11, 14, 15, 13, 12, 16)
 - fewest orders: less than 200,000 (1, 5, 2, 4, 3)

```python
ords_prods_merge.loc [(ords_prods_merge['order_hour_of_day' ] >=10) & (ords_prods_merge['order_hour_of_day']<=16), 'busiest_period_of_day'] = 'Most orders'
```

```python
ords_prods_merge.loc [(ords_prods_merge['order_hour_of_day' ] >=1) & (ords_prods_merge['order_hour_of_day']<=5), 'busiest_period_of_day'] = 'Fewest orders'
```

```python
ords_prods_merge['busiest_period_of_day'].fillna('Average orders', inplace=True)
```
